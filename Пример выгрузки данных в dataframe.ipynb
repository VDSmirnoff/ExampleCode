```python



def fprice(k,lsplit):
    for i in range(len(k)):
        if k[i][0]=='price':
            k[i].append(lsplit[0])
            lsplit.pop(0)
            break;
            
def lencontrol(k):
    for i in range(len(k)):
        if k[i][0]=='price':
            n=len(k[i])
    
    for i in range(len(k)):
        if len(k[i])<n:
            for j in range(n-len(k[i])):
                k[i].append(None)
        if len(k[i])>n:
            print('Ошибка = откуда то взялась лишняя строчка')
            
            
            
def makenewcol(k,newcol,val):
    for i in range(len(k)):
        if k[i][0]=='price':
            n=len(k[i])-1
    k.append([newcol]+[None for i in range(n-1)]+[val])
def colfirst(k,l):
    if len(l.split('$$'))%2==1:
        l = l.replace('$$0$$','$$')####################################
        if len(l.split('$$'))%2==1:
            print('что то не так с четностью типа жилья или года в строке ',l)
    for i in range(0,len(l.split('$$')),2):
        
        f=0
        for j in range(len(k)):
            if k[j][0]==l.split('$$')[i]:
                f=1
                try:
                    k[j].append(l.split('$$')[i+1])
                except Exception:
                    print('Ошибка collfirst')
                    pass
        if f==0:
            makenewcol(k,l.split('$$')[i],l.split('$$')[i+1])

def valfirst(k,l):
    if len(l.split('$$'))%2==1:
        print('что то не так с четностью площади жилья в строке ',l)
    for i in range(0,len(l.split('$$')),2):
        
        f=0
        for j in range(len(k)):
            if k[j][0]==l.split('$$')[i+1]:
                f=1
                try:
                    k[j].append(l.split('$$')[i])
                except Exception:
                    print('Ошибка vallfirst')
                    pass
        if f==0:
            makenewcol(k,l.split('$$')[i+1],l.split('$$')[i])            
            
            
def ftypeyearlift(k,lsplit):
    for i in range(len(lsplit)-1,-1,-1):#обратный отсчет для правильного удаления pop
        if lsplit[i].split('$$')[0] in ['Тип жилья' , 'Год постройки','Лифты','Тип дома','геоданные','адрес','Строительная серия','Подъезды','Тип перекрытий','Парковка','Мусоропровод','Аварийность','аналитика цены','ссылка','изменениецены','месяц']:
            colfirst(k,lsplit[i])
            lsplit.pop(i)
    #for l in lsplit:
     #   if l.split('$$')[0] in ['Тип жилья' , 'Год постройки']:
      #      colfirst(k,l)
    
def fsqr(k,lsplit):
    for i in range(len(lsplit)-1,-1,-1):#обратный отсчет для правильного удаления pop
        zerosplit = lsplit[i].split('$$')[0] 
        if ('м²' in zerosplit) and not('комн' in zerosplit or 'Студия' in zerosplit or 'студия' in zerosplit or 'свободной планировки' in zerosplit):
            valfirst(k,lsplit[i])
            lsplit.pop(i)
            
def adr(k,lsplit):
    q=0
    for i in range(len(lsplit)-1,-1,-1):#обратный отсчет для правильного удаления pop
        zerosplit =  lsplit[i].split('$$')[0]
        if ('комн' in zerosplit) or ('Студия' in zerosplit) or ('студия' in zerosplit) or ('свободной планировки' in zerosplit) :
            q=q+1
            lsplit.pop(i)
    if q>1:print('строк, содержащих КОМН. больше 1')
def remove(lsplit,a):
    try:
        lsplit.remove(a)
    except Exception:
        pass
    
def readfile(file,mass):
    with open(file, 'r',encoding='utf-8') as file1:
        for line in file1:
            if len(line)>10:
                mass.append(line)
def tryreadfile(file,mass):
    try:
        with open(file, 'r',encoding='utf-8') as file1:
            for line in file1:
                if len(line)>10:
                    mass.append(line)
    except Exception:
        print(file,' cant read')
```

```python
def getnumber(df,feature):#берет первую цифру в строке
    z=[]
    for i in range(len(df)):
        try:
            z.append(float(df[feature][i].split(' ')[0].replace(',','.')))
        except Exception:
            if df[feature][i]!= None:
                print('Error: ', df[feature][i])
            z.append(None)
            pass
    df[feature]=z
```

```python
for j in ['Общая','Жилая','Кухня','Высота потолков','месяц']:
    getnumber(df,j)
```

```python
def getsum(df,feature):
    n=0
    z1=[]
    
    for i in range(len(df)):
        nsum=0
        number=''
        if df[feature][i]!= None:
            
            for j in (df[feature][i]).replace(',','.'):

                try:

                    float(number+j)
                    number+=j
                except Exception:
                    if number!='':
                        nsum+=float(number)
                        number=''
                    pass
        if nsum==0:
            z1.append(None)
        else:
            z1.append(round(nsum,4))
    df['Сумма площадей комнат']=z1
def getplusminus(df,feature):
    z1=[];z2=[] 
    for i in range(len(df)):
        if df[feature][i]!= None:
            
            if '+' in df[feature][i]:
                z1.append(df[feature][i].count('+'))
            else:
                z1.append(0)
                
            if '-' in df[feature][i]:
                z2.append(df[feature][i].count('-'))
            else:
                z2.append(0)
        else:
            z1.append(0)
            z2.append(0)
    df['Колво смеж комн']=z1
    df['Колво изолир комн']=z2
```

```python
def getfirstnum(string):
    f=False
    n=''
    for i in string:
        if n!='' and not f:
            break
        try:
            int(i)
            f=True
            n=n+i
        except Exception:
            f=False
            pass
    if n=='' and not f : return('0')#если есть нужное слово но нет цифры
    else: return int(n)
def dual(df,columns, variant):
    
    for c in range(len(columns)):
        nerror=0
        z0=[]
        z1=[]
        zsum=[]
    
        for i in df[columns[c]]:
            if i == 'Нет':
                z0.append(0)
                z1.append(0)
                zsum.append(0)
            elif not (i == i) or i==None:
                z0.append(None)
                z1.append(None)
                zsum.append(None)
                
                
            else:
                z00=False
                z11=False
                for j in i.split(','):
                    try:
                        if variant[c][0] in j:
                            z00=True
                            z0.append(getfirstnum(j))

                        if variant[c][1] in j:
                            z11=True
                            z1.append(getfirstnum(j))

                    except Exception:
                        print(j)
                
                if not z00 and not z11:
                    z0.append(-1)
                    z1.append(-1)
                    
                    print('В ',c,' нет ни ', variant[c][0],' ни ',variant[c][1],' :', i)
                    nerror+=1
                    if getfirstnum(i)=='0':
                        zsum.append(-1)
                    else:
                        zsum.append(getfirstnum(i))
                else:
                    if not z00:
                        z0.append(0)
                    if not z11:
                        z1.append(0)
                    zsum.append(int(z0[len(z0)-1])+int(z1[len(z1)-1]))
                
                    
        df = df.drop(columns[c] , axis=1)
        df[columns[c]+variant[c][0]] = z0
        df[columns[c]+variant[c][1]] = z1
        if columns[c]=='Лифты':
            df[columns[c]+'сумма'] = zsum
        print('                         итого',nerror)
    return df
```
