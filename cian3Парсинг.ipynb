```python
def fprice(k,lsplit):
    for i in range(len(k)):
        if k[i][0]=='price':
            k[i].append(lsplit[0])
            lsplit.pop(0)
            break;
            
def lencontrol(k):
    for i in range(len(k)):
        if k[i][0]=='price':
            n=len(k[i])
    
    for i in range(len(k)):
        if len(k[i])<n:
            for j in range(n-len(k[i])):
                k[i].append(None)
        if len(k[i])>n:
            print('Ошибка = откуда то взялась лишняя строчка')
            
            
            
def makenewcol(k,newcol,val):
    for i in range(len(k)):
        if k[i][0]=='price':
            n=len(k[i])-1
    k.append([newcol]+[None for i in range(n-1)]+[val])
def colfirst(k,l):
    if len(l.split('$$'))%2==1:
        print('что то не так с четностью типа жилья или года в строке ',l)
    for i in range(0,len(l.split('$$')),2):
        
        f=0
        for j in range(len(k)):
            if k[j][0]==l.split('$$')[i]:
                f=1
                try:
                    k[j].append(l.split('$$')[i+1])
                except Exception:
                    pass
        if f==0:
            makenewcol(k,l.split('$$')[i],l.split('$$')[i+1])

def valfirst(k,l):
    if len(l.split('$$'))%2==1:
        print('что то не так с четностью площади жилья в строке ',l)
    for i in range(0,len(l.split('$$')),2):
        
        f=0
        for j in range(len(k)):
            if k[j][0]==l.split('$$')[i+1]:
                f=1
                try:
                    k[j].append(l.split('$$')[i])
                except Exception:
                    pass
        if f==0:
            makenewcol(k,l.split('$$')[i+1],l.split('$$')[i])            
            
            
def ftypeyearlift(k,lsplit):
    for i in range(len(lsplit)-1,-1,-1):#обратный отсчет для правильного удаления pop
        if lsplit[i].split('$$')[0] in ['Тип жилья' , 'Год постройки','Лифты','Тип дома','геоданные','адрес','Строительная серия','Подъезды','Тип перекрытий','аналитика цены','ссылка','изменениецены']:
            colfirst(k,lsplit[i])
            lsplit.pop(i)
    #for l in lsplit:
     #   if l.split('$$')[0] in ['Тип жилья' , 'Год постройки']:
      #      colfirst(k,l)
    
def fsqr(k,lsplit):
    for i in range(len(lsplit)-1,-1,-1):#обратный отсчет для правильного удаления pop
        if ('м²' in lsplit[i].split('$$')[0]) and not('комн' in lsplit[i].split('$$')[0]) and not('свободной планировки' in lsplit[i].split('$$')[0]) and not('тудия' in lsplit[i].split('$$')[0]):
            valfirst(k,lsplit[i])
            lsplit.pop(i)
            
def adr(k,lsplit):
    q=0
    for i in range(len(lsplit)-1,-1,-1):#обратный отсчет для правильного удаления pop
        if ('комн' in lsplit[i].split('$$')[0]) or ('свободной планировки' in lsplit[i].split('$$')[0])  or ('тудия' in lsplit[i].split('$$')[0]):
            q=q+1
            lsplit.pop(i)
    if q>1:print('строк, содержащих КОМН. больше 1')
```
